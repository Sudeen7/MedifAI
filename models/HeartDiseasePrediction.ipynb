```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = SVC(kernel='linear')
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
conf_matrix = confusion_matrix(Y_train, X_train_prediction)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
print("Confusion Matrix for Training Data:")
print(conf_matrix)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
conf_matrix = confusion_matrix(Y_test, X_test_prediction)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
print("Confusion Matrix for Training Data:")
print(conf_matrix)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
heart_data = pd.read_csv('heart.csv')

# Prepare data
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Create pipeline
heart_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(kernel='linear'))
])

# Train model
heart_pipeline.fit(X_train, Y_train)

# Save the pipeline using joblib
joblib.dump(heart_pipeline, 'heart_disease_pipeline.joblib')

print("Pipeline trained and saved successfully.")
```
