```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import make_classification
```

```python
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']

# ecoding prognonsis
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)

# # Generate synthetic dataset for demonstration purposes
# X, y = make_classification(
#     n_samples=1000, n_features=20, n_informative=15, n_classes=3, random_state=42
# )

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
import numpy as np

# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=23),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
}

# Loop through the models, train, test, and print results
for model_name, model in models.items():
    print(f"Training {model_name}...")
    try:
        # Train the model
        model.fit(X_train, y_train)

        # Training set predictions and accuracy
        train_predictions = model.predict(X_train)
        train_accuracy = accuracy_score(y_train, train_predictions)

        # Test set predictions and accuracy
        test_predictions = model.predict(X_test)
        test_accuracy = accuracy_score(y_test, test_predictions)

        # Confusion Matrix
        cm = confusion_matrix(y_test, test_predictions)

        # F1 Score, Precision, Recall
        f1 = f1_score(y_test, test_predictions, average='weighted')
        precision = precision_score(y_test, test_predictions, average='weighted')
        recall = recall_score(y_test, test_predictions, average='weighted')

        # Print results
        print(f"{model_name} Training Accuracy: {train_accuracy:.2f}")
        print(f"{model_name} Testing Accuracy: {test_accuracy:.2f}")
        print(f"{model_name} Confusion Matrix:\n{np.array2string(cm, separator=', ')}")
        print(f"{model_name} F1 Score: {f1:.2f}")
        print(f"{model_name} Precision: {precision:.2f}")
        print(f"{model_name} Recall: {recall:.2f}")

    except Exception as e:
        print(f"Error with {model_name}: {e}")

    print("\n" + "="*40 + "\n")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, Y, cv=5)
print(f"Cross-Validation Scores: {scores}")
print(f"Mean Accuracy: {scores.mean()}")
```

```python
from sklearn.ensemble import VotingClassifier

# Create a VotingClassifier with the trained models
voting_clf = VotingClassifier(estimators=[
    ('SVC', models['SVC']),
    ('RandomForest', models['RandomForest']),
    ('KNeighbors', models['KNeighbors']),
], voting='hard')

# Train VotingClassifier on the same training data
voting_clf.fit(X_train, y_train)

# Test the ensemble
voting_predictions = voting_clf.predict(X_test)
accuracy = accuracy_score(y_test, voting_predictions)
cm = confusion_matrix(y_test, voting_predictions)

print(f"Voting Classifier Accuracy: {accuracy}")
print("Voting Classifier Confusion Matrix:")
print(np.array2string(cm, separator=', '))
```

```python
# Save the trained VotingClassifier to a file
import pickle
with open('voting_clf.pkl', 'wb') as f:
    pickle.dump(voting_clf, f)
```

```python
# Load the VotingClassifier from the file
with open('voting_clf.pkl', 'rb') as f:
    voting_clf = pickle.load(f)

# Use the loaded model to make predictions
voting_predictions = voting_clf.predict(X_test)
accuracy = accuracy_score(y_test, voting_predictions)

print(f"Loaded Voting Classifier Accuracy: {accuracy}")
```
